```python
# importing the necessary libraries.
from bs4 import BeautifulSoup
import requests
import urllib
import pandas as pd
```

```python
# accessing the website.
search_url =  "https://www.getguru.com/customers"
r = requests.get(search_url)
soup = BeautifulSoup(r.text, 'html.parser')
blogs = soup.find_all('div', class_ = 'resource-block w-dyn-item')
images = soup.find_all('img', class_ = 'alternative-card-logo on-card')
```

```python
# finding the images in the links of the blogs.
for i in images:
    print(i['src'])
```

```python
# finding the links of the list of companies.
links =[]
for blog in blogs:
    blog_link = blog.a['href']
    print(f'blog_link : {blog_link}')
    links.append(blog_link) 
```

```python
# finding the names of the companies.
name = []
for link in links:
    s = "https://www.getguru.com/"+link
    r = requests.get(s)
    soup = BeautifulSoup(r.text, 'html.parser')
    title = soup.find('p')
    name.append(title.text)
```

```python
# seperating the names from the list.
customer_name = []
for i in name:
    if i.split()[0] not in customer_name:
        customer_name.append(i.split()[0])
    
```

```python
# creating a dataframe with the customer names.
df_customer = pd.DataFrame(customer_name, columns = ['customer_name'])
df_customer.to_excel('customer_name.xlsx', index=False)
df_customer
```
